```python
import pandas as pd

df = pd.read_excel("ONS.xlsx", sheet_name="MYE4", skiprows=7, header=0)
```

```python
df = df.dropna(axis=1, how="all")
df = df.dropna(axis=0, how="all")
```

```python
print("Column names:")
print(df.columns.tolist())
```

```python
keep = [
    "UNITED KINGDOM",
    "GREAT BRITAIN",
    "ENGLAND",
    "LONDON",
    "Westminster",
    "Camden",
    "Brent",
    "Kensington and Chelsea"
]

df_clean = df[df["Name"].isin(keep)]
```

```python
df_clean.to_csv("data_clean/MYE4_clean.csv", index=False)
```

```python
import pandas as pd

df1 = pd.read_excel("ONS.xlsx", sheet_name="MYE2 - Persons", skiprows=7, header=0)
```

```python
df1 = df1.dropna(axis=1, how="all")
df1 = df1.dropna(axis=0, how="all")
```

```python
keep_rows = [
    "UNITED KINGDOM",
    "ENGLAND",
    "LONDON",
    "Westminster",
    "Camden",
    "Brent",
    "Kensington and Chelsea"
]

df1 = df1[df1["Name"].isin(keep_rows)]
```

```python
age_cols = [col for col in df1.columns if col.isdigit() and int(col) >= 50] + ["90+"]

df_50plus = df1[["Code", "Name", "Geography"] + age_cols]
```

```python

df_50plus.to_csv("data_clean/MYE2_50plus_clean.csv", index=False)
```

```python
import pandas as pd

df2 = pd.read_excel("ONS.xlsx", sheet_name="MYE1", skiprows=7, header=0)
```

```python
df2 = df2.dropna(axis=1, how="all")
df2 = df2.dropna(axis=0, how="all")
```

```python
# Filter for 50+ age groups
age_50plus = [
    "50 to 54", "55 to 59", "60 to 64", "65 to 69", "70 to 74",
    "75 to 79", "80 to 84", "85 to 89", "90+"
]

df2 = df2[df2["Country Code"].isin(age_50plus)]

# Keep only United Kingdom and England columns
df2 = df2[["Country Code", "K02000001", "E92000001"]]

# Rename columns for clarity
df2 = df2.rename(columns={
    "Country Code": "Age Group",
    "K02000001": "United Kingdom",
    "E92000001": "England"
})
```

```python
print("Filtered df2 (50+ age groups, UK and England):")
print(df2)
```

```python
df2.to_csv("data_clean/MYE1_50plus_UK_England.csv", index=False)
```
