```python
import pandas as pd

df = pd.read_excel("ONS.xlsx", sheet_name="MYE4", skiprows=7, header=0)
```

```python
df = df.dropna(axis=1, how="all")
df = df.dropna(axis=0, how="all")
```

```python
print("Column names:")
print(df.columns.tolist())
```

```python
keep = [
    "UNITED KINGDOM",
    "GREAT BRITAIN",
    "ENGLAND",
    "LONDON",
    "Westminster",
    "Camden",
    "Brent",
    "Kensington and Chelsea"
]

df_clean = df[df["Name"].isin(keep)]
```

```python
df_clean.to_csv("data_clean/MYE4_clean.csv", index=False)
```
