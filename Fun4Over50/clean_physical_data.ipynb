```python
import pandas as pd
df = pd.read_excel("data/Physical Inactivity by age and gender.xlsx")
```

```python
print("Physical Inactivity Age Groups:")
print(df.columns.tolist())
print(df.head())
```

```python
df1 = pd.read_excel("data/Self-Reported Loneliness by age.xlsx")
```

```python
print("Loneliness Age Groups:")
print(df1.columns.tolist())
print(df1.head())
```

```python
age_mapping = {
    '45-53': '45-54',
    '55-63': '55-64',
    '65-73': '65-74',  # adjust this if your loneliness data has a different range
    '75+': '75+'
}

df['Age group'] = df['Age group'].replace(age_mapping)
```

```python
print(df['Age group'].unique())
```
