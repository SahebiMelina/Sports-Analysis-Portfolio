```python
import sqlite3
import pandas as pd

df = pd.read_csv("../data/cleaned_athletes.csv")

conn = sqlite3.connect("../data/crossfit.db")
df.to_sql("athletes", conn, if_exists="replace", index=False)

print("Data loaded into SQLite!")
```
