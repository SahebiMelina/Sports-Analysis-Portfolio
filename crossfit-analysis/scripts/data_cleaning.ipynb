```python
import pandas as pd

# Load data
df = pd.read_csv("../data/athletes.csv")

# Standardize column names so that they look the same
df.columns = df.columns.str.lower()

# Rename for a better understanding
df = df.rename(columns={
    "weight": "bodyweight"
})

# Convert numeric columns
numeric_cols = ["age", "height", "bodyweight", "snatch", "deadlift", "backsq", "candj"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Drop missing key values
df = df.dropna(subset=["bodyweight", "snatch"])

# Save cleaned dataset
df.to_csv("../data/cleaned_athletes.csv", index=False)

```

```python
columns_to_keep = [
    "name",
    "age",
    "height",
    "bodyweight",
    "snatch",
    "deadlift",
    "backsq",
    "candj"
]

df_small = df[columns_to_keep]

# 2. Reduce number of rows (choose ONE option)
df_small = df_small.sort_values(by="snatch", ascending=False).head(3000)

# 3. Round numbers (cleaner look)
df_small["height"] = df_small["height"].round(1)
df_small["bodyweight"] = df_small["bodyweight"].round(1)

# 4. Save GitHub-friendly dataset
df_small.to_csv("../data/cleaned_athletes.csv", index=False)

print("✅ Full dataset + GitHub-friendly dataset saved!")
```
