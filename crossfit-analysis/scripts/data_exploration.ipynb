```python
import pandas as pd
df = pd.read_csv("../data/athletes.csv")
print(df.columns)
```

```python
df[['age','train', 'schedule', 'howlong', 'experience']].head(10)
```

```python
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1})
```

```python
df['experience'].unique()
```

```python
def map_experience(x):
    if pd.isna(x):
        return None
    if "1-2" in x:
        return 1.5
    elif "2-4" in x:
        return 3
    elif "4+" in x:
        return 5
    return None

df['howlong_clean'] = df['howlong'].apply(map_experience)
```

```python
def map_schedule(x):
    if pd.isna(x):
        return None
    if "multiple workouts in a day 3+" in x:
        return 3
    elif "multiple workouts in a day 2x" in x:
        return 2
    elif "1 workout a day" in x:
        return 1
    return None

df['schedule_clean'] = df['schedule'].apply(map_schedule)
```

```python
df['train'] = df['train'].fillna('')
train_dummies = df['train'].str.get_dummies(sep='|')
df = pd.concat([df, train_dummies], axis=1)
```

```python
df_selected = df[[ 'name', 'age', 'gender','experience', 'train','schedule','howlong', 'fran']]

df_selected.to_csv('../data/Analysis.csv', index=False)
```
